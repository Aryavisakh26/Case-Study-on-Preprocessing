```python
columns_to_check=['Sale Price','No of Bedrooms','No of Bathrooms','Flat Area (in Sqft)','Lot Area (in Sqft)','Overall Grade','Area of the House from Basement (in Sqft)','Basement Area (in Sqft)','Renovated Year','Latitude','Longitude','Living Area after Renovation (in Sqft)','Lot Area after Renovation (in Sqft)']
```

```python
#Outlier Removal
```

```python
for col in columns_to_check:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR =Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    df=df[(df[col]>=lower_bound) &(df[col]<=upper_bound)]
```
